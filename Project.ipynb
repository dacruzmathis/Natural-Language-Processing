- Replace all non letter characters in the text column with a whitespace and add it as a new column text_clean in the speech_df DataFrame.


```python
import re
```

```python
pattern = re.compile('[^a-zA-Z]')
```

```python
speech_df['text_clean'] = speech_df['text'].apply(lambda x: re.sub(pattern, ' ', x))
```

```python
speech_df['text_clean'] = speech_df['text_clean'].apply(lambda x: re.sub(' +', ' ', x))
```

- Make all characters in the newly created text_clean column lower case.


```python
speech_df['text_clean'] = speech_df['text_clean'].apply(lambda x: x.lower().strip())
```

- Print the first 5 rows of the text_clean column.


```python
speech_df['text_clean'].head()
```


- Record the character length of each speech (speech_df['text_clean']) and store it in a new char_count column.


```python
speech_df['char_count'] = speech_df['text_clean'].apply(lambda x: len(x))
```

```python
speech_df['char_count'].head()
```

- Record the word count of each speech in the word_count column.


```python
speech_df['word_count'] = speech_df['text_clean'].apply(lambda x: len(x.split()))
```

```python
speech_df['word_count'].head()
```

- Record the average word length of each speech in the avg_word_length column.


```python
speech_df['avg_word_length'] = (speech_df['char_count'] - speech_df['text_clean'].apply(lambda x: x.count(" "))) / speech_df['word_count']
```

```python
import emoji
```

Mots qui ne sont pas des emojis ni des #, ni des @ et ni des url

```python
tweets_without = []
for tweet in tweets:
    without_hashtags_url = []
    for word in tweet.split(" "): 
        if '#' not in word and 'http' not in word and '@' not in word and not emoji.emoji_count(word):
            without_hashtags_url.append(word)
    tweet = " ".join(without_hashtags_url)
    if tweet != '':
        tweets_without.append(tweet)
tweets_without
```

```python
import nltk
```

```python
import string
```

```python
tags = '...-_`#@.():«»,?!;"\''
```

Lettre n'etant pas une ponctuation ni un chiffre

```python
clean_tweets = []
for tweet in tweets_without:
    sentence = []
    for letter in nltk.word_tokenize(tweet):
        #if letter not in string.punctuation:
        if letter not in tags and not letter.isdigit():
            sentence.append(letter)
    clean_tweets.append(" ".join(sentence))
clean_tweets
```

```python
import spacy
from spacy.lang.fr import French
from spacy.lang.fr.stop_words import STOP_WORDS
```

```python
import re
```

```python
nlp = French()
```

Mot n'etant pas un stopword, composé uniquement de lettres, composé d'au moins 2 lettres. On met le tout en minuscule

```python
without_stopwords = []
for tweet in clean_tweets:
    filtered_sentence = []
    for word in tweet.split(" "): 
        lexeme = nlp.vocab[word]
        if lexeme.is_stop == False and re.match("^[a-zA-Z]", word) and len(word) > 1:
            filtered_sentence.append(word.lower())
    without_stopwords.append(" ".join(filtered_sentence))
without_stopwords
```

```python
nlp = spacy.load('fr_core_news_sm')
```

```python
tweets_lemmatize = []
for tweet in without_stopwords:
    doc = nlp(tweet)
    tweets_lemmatize.append(' '.join([word.lemma_ for word in doc]))
tweets_lemmatize
```

- Make a sentimental analysis around the "reforme des retraites".


```python
from vaderSentiment_fr.vaderSentiment import SentimentIntensityAnalyzer
```

```python
SIA = SentimentIntensityAnalyzer()
```

```python
pos = []
neu = []
neg = []
count = 0
for tweet in without_stopwords:
    pos.append(SIA.polarity_scores(tweet)['pos'])
    neu.append(SIA.polarity_scores(tweet)['neu'])
    neg.append(SIA.polarity_scores(tweet)['neg'])
    count += 1
```

```python
sum(pos) / count
```

```python
sum(neu) / count
```

```python
sum(neg) / count
```

```python
pos3 = 0
neu3 = 0
neg3 = 0
count3 = 0
for tweet in without_stopwords:
    maxi = max(SIA.polarity_scores(tweet)['pos'], max(SIA.polarity_scores(tweet)['neg'], SIA.polarity_scores(tweet)['neu']))
    if maxi == SIA.polarity_scores(tweet)['pos']: 
        pos3 += 1
    elif maxi == SIA.polarity_scores(tweet)['neg']: 
        neg3 += 1
    else:
        neu3 += 1
    count3 += 1
```

```python
pos3 / count3
```

```python
neg3 / count3
```

- Define a rate of engagement of the followers towards this user (the more the rate increases the more the user is influential)

```python
import pandas as pd
```

```python
def formula(total_interactions, moy_nb_viewers_per_tweet):    
    engagement_rate = total_interactions / df.shape[0]
    engagement_rate_viewers = engagement_rate / moy_nb_viewers_per_tweet
    engagement_rate_followers = engagement_rate / df['user'][0]['followersCount']
    print(f"Nombre d'interactions moyenne par tweets: {int(engagement_rate)}")
    print(f"Taux d'engagement par follower: {engagement_rate_followers}%")
    print(f"Taux d'engagement par viewer: {engagement_rate_viewers}%")
```

```python
def statitics(df):
    df = df [['user', 'replyCount', 'retweetCount', 'likeCount', 'quoteCount' , 'viewCount', 'content']]
    total_interactions = df['quoteCount'].sum() + df['replyCount'].sum() + df['retweetCount'].sum() + df['likeCount'].sum()
    total_interactions_with_weights = (df['replyCount'].sum() * 0.1) + (df['quoteCount'].sum() + df['retweetCount'].sum()) * 0.7 + (df['likeCount'].sum() * 0.2)
    total_followers = df['user'][0]['followersCount']
    total_viewers = int(df['viewCount'].sum())
    moy_nb_viewers_per_tweet = total_viewers / df.shape[0] 
    print(f"Nombre total de followers: {total_followers}")
    print(f"Nombre total de visualisations: {total_viewers}")
    print(f"Nombre de viewers moyens par tweets: {int(moy_nb_viewers_per_tweet)}")
    print('\nNormal formula')
    formula(total_interactions, moy_nb_viewers_per_tweet)
    print('\nWithout weights formula')
    formula(total_interactions_with_weights, moy_nb_viewers_per_tweet)
```

```python
def statitics2(df):
    df = df [['user', 'replyCount', 'retweetCount', 'likeCount', 'quoteCount' , 'viewCount', 'content']]
    moyenne_interactions = round(df['quoteCount'].mean() + df['replyCount'].mean() + df['retweetCount'].mean() + df['likeCount'].mean(), 2)
    moyenne_interactions_with_weights = round((df['quoteCount'].mean() * 0.1) + (df['replyCount'].mean() + df['retweetCount'].mean()) * 0.7 + (df['likeCount'].mean() * 0.2), 2)
    total_followers = df['user'][0]['followersCount']
    moyenne_viewers = int(df['viewCount'].mean())
    print(f"Nombre total de followers: {total_followers}")
    print(f"Nombre moyen de viewers: {moyenne_viewers}")
    print(f"Nombre moyens d'interactions: {moyenne_interactions}")
    print('\nNormal formula')
    formula2(moyenne_interactions, moyenne_viewers)
    print('\nWithout weights formula')
    formula2(moyenne_interactions_with_weights, moyenne_viewers)
```

```python
def formula2(moyenne_interactions, moyenne_viewers):    
    engagement_rate_viewers = round((moyenne_interactions / moyenne_viewers) * 100, 2)
    engagement_rate_followers = round(moyenne_interactions / df['user'][0]['followersCount'], 2)
    print(f"Taux d'engagement par follower: {engagement_rate_followers}%")
    print(f"Taux d'engagement par viewer: {engagement_rate_viewers}%")
```
